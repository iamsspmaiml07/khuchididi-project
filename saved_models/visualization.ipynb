```python
import os
import pickle
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
if not os.path.exists('visualizations'):
    os.makedirs('visualizations')
```

```python
brain=pd.read_csv('Brain Tumor.csv')
corona=pd.read_csv('Corona.csv')
diabetes=pd.read_csv('diabetes.csv')
heart=pd.read_csv('heart.csv')
head=pd.read_csv('parkinsons.csv')
```

```python
#Brain Tumor
# Plot a breakdown of patient outcomes with a title
plt.figure(figsize=(8, 6))
plt.title("Breakdown of Patient Outcomes in Brain Tumor Cases", fontsize=14, fontweight='bold')
brain["Outcome"].value_counts().plot.pie(autopct="%1.1f%%", colors=sns.color_palette("coolwarm", 3))
plt.ylabel("")
plt.savefig('visualizations/brain1.png')
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=brain, x="Outcome", y="Tumor Size (cm)", palette="pastel")
median_tumor_size = brain.groupby("Outcome")["Tumor Size (cm)"].median().reset_index()
sns.lineplot(data=median_tumor_size, x="Outcome", y="Tumor Size (cm)", marker="o", color="red", label="Median Tumor Size")
plt.title("Tumor Size Distribution by Outcome", fontsize=14, fontweight='bold')
plt.xlabel("Patient Outcome", fontsize=12)
plt.ylabel("Tumor Size (cm)", fontsize=12)
plt.legend()
plt.savefig('visualizations/brain3.png')
```

```python
symptoms = ['Cough', 'Sore Throat', 'Chest Pain', 'Taste Loss', 'Fatigue Level']
positive_cases = corona[corona['COVID Positive'] == 'Yes']
symptom_counts = positive_cases[symptoms].apply(lambda x: (x == 'Yes').sum())
plt.figure(figsize=(8,5))
sns.barplot(x=symptom_counts.index, y=symptom_counts.values, palette='Reds_r')
plt.title("Most Common Symptoms in COVID Positive Cases")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.savefig('visualizations/covid2.png')
```

```python
diabetes_counts = diabetes['Outcome'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(diabetes_counts, labels=['Non-Diabetic', 'Diabetic'], autopct='%1.1f%%', colors=['lightblue', 'salmon'])
plt.title("Diabetes Prevalence Rate")
plt.savefig('visualizations/diabetes2.png')
```

```python
plt.figure(figsize=(8,5))
sns.histplot(diabetes[diabetes['Outcome']==1]['Age'], bins=30, kde=True, color='red')
plt.title("Age Distribution of Diabetic Patients")
plt.xlabel("Age")
plt.ylabel("Count")
plt.savefig('visualizations/diabetes1.png')
```

```python
plt.figure(figsize=(8,5))
heart['target'].value_counts().plot.pie(autopct='%1.1f%%', labels=['No Disease', 'Disease'], colors=['skyblue', 'salmon'])
plt.title("Heart Disease Distribution")
plt.savefig('visualizations/heart1.png')
```

```python
sns.countplot(x=heart['cp'], hue=heart['target'], palette=['blue', 'red'])
plt.title("Chest Pain Type Distribution")
plt.xlabel("Chest Pain Type")
plt.ylabel("Count")
plt.legend(title="Heart Disease", labels=["No Disease (0)", "Disease (1)"])
plt.savefig('visualizations/heart2.png')
```

```python
head['status'].value_counts().plot.pie(autopct='%1.1f%%', labels=['Healthy', 'Parkinson’s'], colors=['skyblue', 'salmon'])
plt.title("Parkinson’s Disease Distribution")
plt.savefig('visualizations/head1.png')
```

```python
sns.barplot(x=head['status'], y=head['MDVP:Fo(Hz)'], palette=['blue', 'red'])
plt.title("Mean Fundamental Frequency by Status")
plt.xlabel("Status (0=Healthy, 1=Parkinson’s)")
plt.ylabel("MDVP:Fo (Hz)")
plt.savefig('visualizations/head2.png')
```
